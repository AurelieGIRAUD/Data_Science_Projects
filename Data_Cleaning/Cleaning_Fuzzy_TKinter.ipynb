```python
import pandas as pd 
import numpy as np
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import tkinter as tk
from functools import partial
from collections import OrderedDict
```

## Match names

```python
df = pd.read_csv('basket_of_techs.csv')
df = df.drop('NaN', axis=1)
df.set_index(['name','sum'], inplace= True)
df
```

```python
# Get list of column names
word_list = df.columns.to_list()

# Set a similarity threshold
similarity_threshold = 80

# Group similar words
groups = []
matched_words = set()
group_count = 1

for i, word in enumerate(word_list):
    if word not in matched_words:
        matched_words.add(word)
        group = [word]
        for j, other_word in enumerate(word_list[i + 1:]):
            similarity_score = fuzz.ratio(word, other_word)
            if similarity_score >= similarity_threshold:
                matched_words.add(other_word)
                group.append(other_word)
        groups.append((group_count, group))
        group_count += 1
        
# Create the dictionary with the first word (before any whitespace) of each group as the key and including it in the values
dictionary_of_groups = {group[0].split()[0]: group for _, group in groups}
new_dict = {key: value for key, value in dictionary_of_groups.items() if isinstance(value, list) and len(value) > 1}
ordered_dict = OrderedDict(sorted(new_dict.items()))
```

## Improve the names

```python
# Ask if the user wants to make any changes
choice = input("Do you want to make any changes? (Yes/No): ")
if choice.lower() == 'yes':
    
    # Create a Tkinter window
    window = tk.Tk()
    window.title("Group Selection")
    window.geometry("400x1000")
    
    # Create a canvas for scrolling
    canvas = tk.Canvas(window)
    canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    # Add a scrollbar to the canvas
    scrollbar = tk.Scrollbar(window, command=canvas.yview)
    scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
    canvas.configure(yscrollcommand=scrollbar.set)
    
    # Create a frame inside the canvas to hold the checkboxes
    checkbox_frame = tk.Frame(canvas)
    canvas.create_window((0, 0), window=checkbox_frame, anchor=tk.NW)
    
    # Create variables to track the checkbox states
    checkbox_vars = {}
    selected_groups = {}
    
    def on_checkbox_click(group_number, checkbox_var):
        if checkbox_var.get():
            selected_groups[group_number] = dictionary_of_groups[group_number]
        else:
            selected_groups.pop(group_number, None)


    # Create the checkboxes and populate the dictionary of selected groups
    for group_number, group_values in dictionary_of_groups.items():
        if len(group_values) > 1:
            checkbox_var = tk.IntVar()
            checkbox_vars[group_number] = checkbox_var
            group_text = f"Group {group_number}: {', '.join(group_values)}"
            checkbox = tk.Checkbutton(checkbox_frame, text=group_text, variable=checkbox_var,
                                      command=partial(on_checkbox_click, group_number, checkbox_var))
            checkbox.pack(anchor=tk.W)
            selected_groups[group_number] = group_values  # Initially include all groups in the selected groups dictionary
    
    # Update the canvas scroll region
    canvas.update_idletasks()
    canvas.configure(scrollregion=canvas.bbox("all"))
    
    # Configure the canvas to scroll with the scrollbar
    canvas.bind("<Configure>", lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
    
    # Configure the scrollbar to work with the canvas
    canvas.configure(yscrollcommand=scrollbar.set)
    scrollbar.configure(command=canvas.yview)
    
    # Run the Tkinter event loop
    window.mainloop()
    
    # Prompt the user for new keys 
    new_groups = {}
    for group_number, group_values in selected_groups.items():
        checkbox_var = checkbox_vars[group_number]
        new_key = input(f"Enter the new key for Group {group_number}: ")
        new_groups[new_key] = group_values
        checkbox_var.set(1)  # Set the checkbox as checked
    
    print("Selected Groups:")
    for group_number, group_values in new_groups.items():
        print(f"Group {group_number}: {group_values}")
```

```python
# Merge the 2 dictionaries together to get ALL columns
small_dict= {key: value for key, value in dictionary_of_groups.items() if isinstance(value, list) 
             and len(value) == 1}

print(len(small_dict))
small_dict.update(new_groups)
print(len(small_dict))
```

## Final Dataframe

```python
# Initialize a list to store the summed columns
summed_columns = []

# Iterate over the dictionary items
for key, values in small_dict.items():
    # Get the columns that match the values
    matching_columns = [col for col in df.columns if col in values]
    # Sum the matching columns and append the result to the list
    summed_columns.append(df[matching_columns].sum(axis=1))

# Concatenate the summed columns into a new DataFrame
summed_df = pd.concat(summed_columns, axis=1)

# Assign the dictionary keys as column names
summed_df.columns = small_dict.keys()
```
